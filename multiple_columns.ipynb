```python
import numpy as np
# Multiple aggreagations 

df['A_halved_rounded'] = df['A'].transform(lambda x: x // 2)
df['A_isBig'] = np.where(df['A_halved_rounded'] > 2, 'Big', 'Small')

aggregations = {
    'B': {
      'mean_B': 'mean',
      'min_B': 'min'
    }     
}

def my_agg(x):
    names = {
        'mean_B': x['B'].mean(),
        'mean_C': x['C'].transform(lambda x: x // 2).mean()
    }

    return pd.Series(names, index=['mean_B', 'mean_C'])


df.groupby(['A_halved_rounded', 'A_isBig']).apply(my_agg)
```
